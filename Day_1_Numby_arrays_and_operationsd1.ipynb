```python
import numpy as np
#### Create 1D,2D,3D arrays
array1d=np.array([10,20,30,40])
array2d=np.array([[1,2,3],[4,5,6]])
array3d=np.array([[[1,2],[3,4],[5,6]]])
print("1D Array:",array1d)
print("2D Array:",array2d)
print("3D Array:",array3d)
##### Arithmetic operations
print("\nArithmetic operations")
print("Addition:",array1d +5)
print("subtraction:",array1d -2)
print("Multiplication:",array1d *4)
print("Division:",array1d / 3)
#### Statistical function
print("\nStatistical function")
print("Mean:",np.mean(array1d))
print("Sum:",np.sum(array1d))
print("Standard deviation:",np.std(array1d))
#### Reshape,arange,linspace
reshaped_array = np.arange(20).reshape(4,5) # Corrected np.arrange and chained reshape
print("\nReshape using arange:\n",reshaped_array)
line_array =np.linspace(0,1,3) # Corrected np.linespace
print("linspace(0 to 1):",line_array)
### Index of max,min values
print("\nmax value index:",np.argmax(array1d))
print("\nmin value index:",np.argmin(array1d))
```
