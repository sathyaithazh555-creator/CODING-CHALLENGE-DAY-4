```python
### Creating a 2D list where each row holds a contact's name,phone number,email address
```

```python
contacts = [["depika","7783899292","depikappalani@gmail.com"],
["durga","7657890987","durgasulop@gmail.com"],
["ithazhya","9876543210","ithazhgowtham@mail.com"]            ]
print(type(contacts))
print(contacts)
```

```python
### allow the user to add new contact
```

```python
name =input("Enter the name: ")
phone_number = int(input("Enter the phone_number: "))
email_address = input("Enter the email_address: ")
```

```python
contacts.append([name,phone_number,email_address])
print(contacts)
```

```python
### Display all contacts sorted by name
```

```python
contacts.sort()
print(contacts)
```

```python
### Allow the user to search for contacts by name
```

```python
Name = input("Enter the name: ")
for i in contacts:
  if Name in i:
    print(i)
```
