```python
import pandas as pd

df = pd.read_csv("https://drive.google.com/uc?export=download&id=1GJwDieiqffFzEZtZ3XXVd3jXylHnzFba")


print("\n=== ðŸ“Œ COLUMN NAMES ===")
print(df.columns)

print("\n=== ðŸ“Œ SERIES EXAMPLE (Age) ===")
print(df["Age"].head())

# Add new column
df["Attrition_Flag"] = df["Attrition"].apply(lambda x: 1 if x == "Yes" else 0)

# Update column
df["MonthlyIncome_K"] = df["MonthlyIncome"] / 1000

# Delete unwanted columns if they exist
for col in ["EmployeeCount", "StandardHours", "Over18"]:
    if col in df.columns:
        df.drop(columns=[col], inplace=True)

# Add new row (for demo)
new_row = df.iloc[0].copy()
df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

# Delete last row (demo)
df.drop(df.tail(1).index, inplace=True)

# Set custom index
if "EmployeeNumber" in df.columns:
    df.set_index("EmployeeNumber", inplace=True)

# Sort by MonthlyIncome
df_sorted = df.sort_values(by="MonthlyIncome", ascending=False)

print("\n=== ðŸ“Œ FINAL DATASET INFO ===")
print(df.info())

print("\n=== ðŸ“Œ FIRST 5 ROWS AFTER CHANGES ===")
print(df.head())

print("\n=== ðŸ“Œ TOP 5 BY MONTHLY INCOME ===")
print(df_sorted[["MonthlyIncome", "MonthlyIncome_K", "Attrition", "Attrition_Flag"]].head())

print("\nðŸŽ‰ DAY 2 COMPLETED SUCCESSFULLY â€” NO ERRORS ðŸŽ‰")
```
